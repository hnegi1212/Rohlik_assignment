```python
# Rohlik ML Task — EDA + Forecast/Pricing Visuals (run in a notebook cell)
# Assumes these files are in the current working directory:
# - ml_task_data.csv
# - objective2_forecast.csv
# - objective1_optimal_pricing.csv
# - model_selection.csv

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

DATA_PATH = Path("ml_task_data.csv")
OBJ2_PATH = Path("objective2_forecast.csv")
OBJ1_PATH = Path("objective1_optimal_pricing.csv")
SEL_PATH  = Path("model_selection.csv")

pd.set_option("display.max_columns", 200)
pd.set_option("display.width", 140)

print("Files present:")
for p in [DATA_PATH, OBJ2_PATH, OBJ1_PATH, SEL_PATH]:
    print(f" - {p}: {'OK' if p.exists() else 'MISSING'}")


# Load raw data

df = pd.read_csv(DATA_PATH)
df["date"] = pd.to_datetime(df["date"])
df = df.sort_values(["product_id", "date"]).reset_index(drop=True)

# Derive buy_price from margin definition
df["buy_price"] = df["sell_price"] - df["margin"]

display(df.head())
print("\nRaw data shape:", df.shape)
print("Products:", sorted(df["product_id"].unique()))
print("Date range:", df["date"].min().date(), "→", df["date"].max().date())


# Quick sanity checks

print("\nMissing values per column:")
print(df.isna().sum())

print("\nRanges:")
print("Sales min/max:", df["sales"].min(), df["sales"].max())
print("Sell price min/max:", df["sell_price"].min(), df["sell_price"].max())
print("Margin min/max:", df["margin"].min(), df["margin"].max())
print("Buy price min/max:", df["buy_price"].min(), df["buy_price"].max())


# Helper: plot time series per product in separate figures

def plot_by_product(df_in, y_col, title_prefix, y_label=None):
    for pid, g in df_in.groupby("product_id"):
        g = g.sort_values("date")
        plt.figure()
        plt.plot(g["date"], g[y_col])
        plt.title(f"{title_prefix} — Product {pid}")
        plt.xlabel("Date")
        plt.ylabel(y_label or y_col)
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()


# 1) Raw Data — EDA Charts

plot_by_product(df, "sales", "Daily Sales", "Sales")
plot_by_product(df, "sell_price", "Sell Price", "Sell Price")
plot_by_product(df, "margin", "Margin", "Margin")
plot_by_product(df, "buy_price", "Derived Buy Price", "Buy Price")

# Sales vs price scatter (price sensitivity intuition)
for pid, g in df.groupby("product_id"):
    plt.figure()
    plt.scatter(g["sell_price"], g["sales"], alpha=0.6)
    plt.title(f"Sales vs Sell Price — Product {pid}")
    plt.xlabel("Sell Price")
    plt.ylabel("Sales")
    plt.tight_layout()
    plt.show()

# Weekday seasonality
tmp = df.copy()
tmp["dow"] = tmp["date"].dt.dayofweek  # 0=Mon ... 6=Sun
for pid, g in tmp.groupby("product_id"):
    grp = g.groupby("dow")["sales"].mean()
    plt.figure()
    plt.plot(grp.index, grp.values, marker="o")
    plt.title(f"Avg Sales by Weekday — Product {pid} (0=Mon ... 6=Sun)")
    plt.xlabel("Day of Week")
    plt.ylabel("Avg Sales")
    plt.xticks(range(7))
    plt.tight_layout()
    plt.show()

# Distributions
for col in ["sales", "sell_price", "margin"]:
    plt.figure()
    plt.hist(df[col].values, bins=30)
    plt.title(f"Distribution — {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()


# 2) Predictions & Pricing Outputs

obj2 = pd.read_csv(OBJ2_PATH)
obj1 = pd.read_csv(OBJ1_PATH)
sel  = pd.read_csv(SEL_PATH)

obj2["date"] = pd.to_datetime(obj2["date"])
obj1["date"] = pd.to_datetime(obj1["date"])

display(obj2.head())
display(obj1.head())
print("\nObjective2 rows:", len(obj2), "Objective1 rows:", len(obj1))

# Forecast window
last_date = df["date"].max()
forecast_dates = sorted(obj2["date"].unique())
print("\nLast historical date:", last_date.date())
print("Forecast date range:", forecast_dates[0].date(), "→", forecast_dates[-1].date())

# Historical sales + Obj2 forecast overlay
N_HISTORY = 60  # last 60 days for context
for pid, g in df.groupby("product_id"):
    g = g.sort_values("date")
    hist = g[g["date"] >= (last_date - pd.Timedelta(days=N_HISTORY))]
    f = obj2[obj2["product_id"] == pid].sort_values("date")

    plt.figure()
    plt.plot(hist["date"], hist["sales"], label="Historical Sales")
    plt.plot(f["date"], f["pred_sales"], label="Forecast (Obj2)")
    plt.axvline(last_date, linestyle="--", label="Forecast Start")
    plt.title(f"Sales: History + Obj2 Forecast — Product {pid}")
    plt.xlabel("Date")
    plt.ylabel("Sales")
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Price plan comparison: Obj2 constant vs Obj1 optimized
for pid in sorted(df["product_id"].unique()):
    f2 = obj2[obj2["product_id"] == pid].sort_values("date")
    f1 = obj1[obj1["product_id"] == pid].sort_values("date")

    plt.figure()
    plt.plot(f2["date"], f2["sell_price"], label="Price (Obj2: constant)")
    plt.plot(f1["date"], f1["sell_price"], label="Price (Obj1: optimized)")
    plt.title(f"7-Day Price Plan — Product {pid}")
    plt.xlabel("Date")
    plt.ylabel("Sell Price")
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Aggregate revenue/profit per day: Obj2 vs Obj1
agg2 = obj2.groupby("date")[["expected_revenue", "expected_profit"]].sum().reset_index()
agg1 = obj1.groupby("date")[["expected_revenue", "expected_profit"]].sum().reset_index()

plt.figure()
plt.plot(agg2["date"], agg2["expected_revenue"], label="Obj2 Revenue (constant price)")
plt.plot(agg1["date"], agg1["expected_revenue"], label="Obj1 Revenue (optimized)")
plt.title("Total Expected Revenue by Day")
plt.xlabel("Date")
plt.ylabel("Expected Revenue")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure()
plt.plot(agg2["date"], agg2["expected_profit"], label="Obj2 Profit (constant price)")
plt.plot(agg1["date"], agg1["expected_profit"], label="Obj1 Profit (optimized)")
plt.title("Total Expected Profit by Day")
plt.xlabel("Date")
plt.ylabel("Expected Profit")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Profit constraint: cumulative profit vs threshold
last_rows = df.sort_values("date").groupby("product_id").tail(1)
profit_last_day_total = float((last_rows["margin"] * last_rows["sales"]).sum())
profit_required = 7.0 * profit_last_day_total

cum_profit_obj1 = agg1.sort_values("date")["expected_profit"].cumsum()

plt.figure()
plt.plot(agg1["date"], cum_profit_obj1, label="Cumulative Profit (Obj1)")
plt.axhline(profit_required, linestyle="--", label=f"Required Profit Threshold = {profit_required:,.2f}")
plt.title("Objective 1 — Cumulative Profit vs Required Threshold")
plt.xlabel("Date")
plt.ylabel("Cumulative Expected Profit")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print(f"\nProfit last day (actual): {profit_last_day_total:,.2f}")
print(f"Required 7-day profit:    {profit_required:,.2f}")
print(f"Obj1 7-day profit:        {float(agg1['expected_profit'].sum()):,.2f}")

# Model selection summary charts
sel_obj2 = sel[sel.get("objective").fillna("") == "obj2_forecasting"].copy()
if len(sel_obj2) > 0 and "chosen_forecaster" in sel_obj2.columns:
    plt.figure()
    counts = sel_obj2["chosen_forecaster"].value_counts()
    plt.bar(counts.index.astype(str), counts.values)
    plt.title("Chosen Forecaster Counts (Objective 2)")
    plt.xlabel("Model")
    plt.ylabel("Number of Products")
    plt.tight_layout()
    plt.show()

sel_obj1 = sel[sel.get("objective").fillna("") == "obj1_demand_model"].copy()
if len(sel_obj1) > 0 and "chosen_model" in sel_obj1.columns:
    plt.figure()
    counts = sel_obj1["chosen_model"].value_counts()
    plt.bar(counts.index.astype(str), counts.values)
    plt.title("Chosen Demand Model Counts (Objective 1)")
    plt.xlabel("Model")
    plt.ylabel("Number of Products")
    plt.tight_layout()
    plt.show()

display(sel)
```
